```python
# Data
x = np.arange(0, 10, 1)  # X-axis: 0 to 9
y = x ** 2               # Y-axis: Square of x

# Basic Line Plot
plt.plot(x, y)
```

```python
# Data
x = np.arange(0, 10, 1)  # X-axis: 0 to 9
y = x ** 2               # Y-axis: Square of x

# Basic Line Plot
plt.plot(x, y)

# Adding labels and title
plt.xlabel('X-axis (Numbers)')
plt.ylabel('Y-axis (avarage)')
plt.title('Simple Line Plot for demonstration')

# Display the plot
plt.show()
```

```python
# Data for plotting a simple line chart
x = np.arange(0, 20, 1)  # X-axis: 0 to 9
y = x ** 2               # Y-axis: Square of x

# Basic Line Plot
plt.plot(x, y)

# Adding labels and title
plt.xlabel('X-axis (Numbers)')
plt.ylabel('Y-axis (Squares)')
plt.title('Simple Line Plot')

# Display the plot
plt.show()
# Key Functions:
# plot() â†’ Draws the line chart.
# xlabel(), ylabel() â†’ Labels for the axes.
# title() â†’ Adds a title.
# show() â†’ Displays the plot in Jupyter.
```

```python
# Custom Line Plot
plt.plot(x, y, color='red', linestyle='--', marker='x')

plt.xlabel('x-axis Numbers')
plt.ylabel('Squares')
plt.title('Customized Line Plot demo')

plt.grid(True)  # Adds a grid for better readability
plt.show()

#Customizations:
#color='red' â†’ Changes line color.
#linestyle='--' â†’ Makes the line dashed.
#marker='o' â†’ Adds circular markers to data points.
#grid() â†’ Adds grid lines.
```

```python
# Generating Sample Data for Khubh Mela
days = ['Day 1', 'Day 2', 'Day 3', 'Day 4', 'Day 5','Day 6']
visitors = np.random.randint(50000, 200000, size=6)
cities = ['Prayagraj', 'Varanasi', 'Lucknow', 'Kanpur', 'Agra', 'delhi', 'chandigad']
participants = np.random.randint(3000, 15000, size=7)
visitors
#array([122038, 132960,  78229, 108392, 143408], dtype=int32)
```

```python
# Creating DataFrames
df_visitors = pd.DataFrame({'Day': days, 'Visitors': visitors})
df_participants = pd.DataFrame({'City': cities, 'Participants': participants})
df_visitors
df_participants
```

```python
# Line Plot: Number of Visitors Over Different Days
plt.figure(figsize=(8, 5))
plt.plot(df_visitors['Day'], df_visitors['Visitors'], marker='x', color='green')
plt.xlabel('Days')
plt.ylabel('Number of Visitors')
plt.title('Visitors During Khubh Mela Over Days')
plt.grid(True)
plt.show()
```

```python
# Bar Chart: Participants from Different Cities
plt.figure(figsize=(8, 5))
plt.bar(df_participants['City'], df_participants['Participants'], color='blue')
plt.xlabel('Cities')
plt.ylabel('Number of Participants')
plt.title('Participants from Different Cities - Khubh Mela')
plt.show()
```

```python
# Pie Chart: Distribution of Participants by City
plt.figure(figsize=(8, 8))
plt.pie(df_participants['Participants'], labels=df_participants['City'], autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Participant Distribution by City')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming df_visitors and df_participants are defined correctly.

# Create a figure for multiple subplots
plt.figure(figsize=(12, 6))

# Subplot 1: Line Plot for Daily Visitors
plt.subplot(1, 2, 1)
plt.plot(df_visitors['Day'], df_visitors['Visitors'], marker='s', color='green')
plt.title('Daily Visitors')
plt.xlabel('Days')
plt.ylabel('Visitors')

# Subplot 2: Bar Chart for City-wise Participants
plt.subplot(1, 2, 2)
plt.bar(df_participants['City'], df_participants['Participants'], color='purple')
plt.title('City-wise Participants')
plt.xlabel('Cities')
plt.ylabel('Participants')

# Adjust the layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()

# Adding Legends with Multiple Lines
plt.figure(figsize=(8, 5))

# Ensure you have `visitors` and `participants` data (using df_visitors and df_participants here)
x = np.arange(1, 6)
plt.plot(x, df_visitors['Visitors'].iloc[:5], label='Visitors Trend', linestyle='-', marker='o')  # Adjust based on actual data
plt.plot(x, df_participants['Participants'].iloc[:5], label='Participants Trend', linestyle='--', marker='s')  # Adjust based on actual data

plt.xlabel('Days/Cities Index')
plt.ylabel('Count')
plt.title('Visitors & Participants Trend - Khubh Mela')
plt.legend()
plt.grid(True)

# Show the plot with the legend
plt.show()

# Saving a Plot Example
plt.figure(figsize=(8, 5))
plt.plot(df_visitors['Day'], df_visitors['Visitors'], color='red')
plt.title('Saved Plot Example - Visitors Trend')

# Save the plot as a .png file
plt.savefig('khubh_mela_visitors_trend.png')

# Show the saved plot
plt.show()
```
